```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
df = pd.read_csv('cuisines.csv')
training_data = df.sample(frac=0.8, random_state=25)
testing_data = df.drop(training_data.index)
print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
targets_training = training_data.cuisine
targets_test = testing_data.cuisine
#i dropped the cuisine into a separate array so i can cross reference it later on
training_data.drop(['cuisine'], 1, inplace=True)
testing_data.drop(['cuisine'], 1, inplace=True)

def findDistance(training_row,test_row):
    distance = 0
    for i in range(len(training_row)-1):
		#note because the data starts with the index, i needed to skip that so I do i+1 
        distance += (training_row[i+1] - test_row[i+1])**2 #this takes the difference of each variable and I square it to get the positive value
    return distance**0.5 #this returns the sqrt of the distance this is euclidean distance

def findNeighbors(training_set, test_row, num_neighbors):
	#print(training_set)
	distances = list()#this is a list of tuples that stores the index and the distance for that index. I will match the index to the target array to find the cuisine later
	i = 0
	#loop over all the training examples rows
	for j in range(len(training_set)):
		if j > 0:
			dist = findDistance(training_set.iloc[j], test_row)
			#adds to the distance array
			distances.append((training_set.iloc[j].iloc[0], dist))
		j+=1
	#a method to sort the array from smallest distances to largest distances	
	distances.sort(key=lambda tup: tup[1])

	neighbors = list()
	#this neighbors array is how I find out which food group the model predicts
	for i in range(num_neighbors):
		#i take the index tuple of the distances array
		neighbors.append(distances[i][0])
		
	return neighbors

	

def prediction(list):
    counter = 0
    num = list[0]
     
    for i in list:
		#this method is how I determined the most frequent food group
        curr_frequency = list.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num



def isCorrect(prediction, real_value):
	
	if (prediction == real_value):
		
		global numCorrect
		numCorrect += 1
	else:
		global numIncorrect
		numIncorrect += 1
		
	# print(numCorrect/(numCorrect+numIncorrect))
	# print("correct:", numCorrect, "Total:", numCorrect+numIncorrect)

numCorrect = 0
numIncorrect = 0

#This is my "main method" where all the events of the test set are looped over
for i in range(len(testing_data)):
	#to change the number of neighbors change the last of the input parameters in findNeighbors function
	test = findNeighbors(training_data,testing_data.iloc[i], 7)
	#here test is the array of indexes that correspond to the food groups
	prediction_list = []
	for j in test:
		#print(targets_training[j])
		
		prediction_list.append(targets_training[j])#this line goes through and finds the food group from the index array that is returned from findNeighbors method
	
	#print("the real answer is ", targets_test.iloc[i])	
	#this line predicts if the correct response is given
	isCorrect(prediction(prediction_list), targets_test.iloc[i])


print("correct:", numCorrect, "Total:", numCorrect+numIncorrect)
print("accuracy:", numCorrect/(numCorrect+numIncorrect))
# 0.7571964956195244 for 3 NN

# 0.7609511889862328 for NN 7

# print(findDistance(training_data.iloc[1], testing_data.iloc[1]))
```
